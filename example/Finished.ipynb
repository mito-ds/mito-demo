5. Run the same analysis on Test2.csv, using the code Mito generates

```python
#Import pandas and mitosheet
import pandas as pd
import mitosheet
df = pd.read_csv('./Test1.csv')

# MITO CODE START

df['C'] = df['A'] + df['B']

# MITO CODE END

# Render df as a mitosheet
mitosheet.sheet(df)
```
